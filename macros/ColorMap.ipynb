```python
import matplotlib.pyplot as plt

# Example data points
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange']

# Create scatter plot
fig, ax = plt.subplots()
for i, label in enumerate(labels):
    ax.scatter(x[i], y[i], c=colors[i], label=label, s=400, marker='s')
    ax.annotate(label, (x[i], y[i]), fontsize=10, ha='center', va='center')

# Add color legend
legend = ax.legend(loc='upper left', bbox_to_anchor=(1,1))
for handle in legend.legendHandles:
    handle._sizes = [80]

# Customize plot
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot with Squares and Color Legend')
plt.grid()
plt.show()
```
